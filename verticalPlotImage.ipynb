```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
fileName='run5ShutterOpen'
fitsFile = fits.open(fileName+'.fits') 
imagesArr = fitsFile[0].data
imagesArr=imagesArr.astype(float)
imagesArr=np.flip(imagesArr,axis=1)
trim=1500
imagesArr[imagesArr>trim]=trim
imageSum=np.mean(imagesArr[23:26],axis=0)[150:-100,250:-220]
plt.imshow(imageSum)
plt.show()
y=np.mean(imageSum,axis=1)
x=np.linspace(0,.2*y.shape[0],num=y.shape[0])
plt.plot(x,y)
plt.grid()
plt.show()
```
