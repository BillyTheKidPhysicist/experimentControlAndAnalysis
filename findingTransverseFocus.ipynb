This is a notebook used to find the transverse focus with the moving camera and laser on the platform system. You can find the focus assuming the profile should be gaussian, or more generally use Radial Basis Functions, a powerful method of interpolation that can smoothly fit data well and is technically a type of machine learning.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as spi
```

```python
xUnits='mm' #provide the units as a string
xData=np.asarray([103.0,100,106,109,112,115,118])
yData=np.asarray([243,165,407,422,268,186,114])
plt.scatter(xData,yData)
plt.xlabel(xUnits)
plt.ylabel('signal')
plt.grid()
plt.show()
```

Fit the data with RBF functions

```python
#find max
fitFunc=spi.Rbf(xData,yData,smooth=1e-0) #choose the right value for smooth. start at 1.0 and play with it
xInterp=np.linspace(xData.min(),xData.max(),num=1000)
yInterp=fitFunc(xInterp)
xPeak=xInterp[np.argmax(yInterp)] #find the peak, this does not work with multiple peaks
plt.title('Peak is at '+str(np.round(xPeak,1))+' '+xUnits)
plt.scatter(xData,yData)
plt.plot(xInterp,yInterp,c='r')
plt.xlabel(xUnits)
plt.ylabel('signal')
plt.axvline(x=xPeak,c='orange')
plt.grid()
plt.show()

```

```python
#find min
fitFunc=spi.Rbf(xData,yData,smooth=1e-1) #choose the right value for smooth. start at 1.0 and play with it
xInterp=np.linspace(xData.min(),xData.max(),num=1000)
yInterp=fitFunc(xInterp)
xPeak=xInterp[np.argmax(yInterp)] #find the peak, this does not work with multiple peaks
plt.title('Peak is at '+str(np.round(xPeak,1))+' '+xUnits)
plt.scatter(xData,yData)
plt.plot(xInterp,yInterp,c='r')
plt.xlabel(xUnits)
plt.ylabel('signal')
plt.axvline(x=xPeak,c='orange')
plt.grid()
plt.show()

```
