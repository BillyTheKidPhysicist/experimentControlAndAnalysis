```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as spi
import scipy.optimize as spo

#FWHM
xData=np.asarray([40,90,120,150,130,110,60,105])
#xMean=np.mean(xData)
#xData=xData-xMean
#xData=-xData+xMean
yData=np.asarray([5.3,4.74,4.66,5.15,5.0,4.865,5.16,4.969])

def fit(x,x0,a,b,c):
  return b+a*np.sqrt((x-x0)**2+c**2)


guess=[100,.01,5.0,50]
bounds=[(-np.inf,0,0,0),(np.inf,np.inf,np.inf,np.inf)]
params,pcov=spo.curve_fit(fit,xData,yData,bounds=bounds,p0=guess,maxfev=100000)
print(params)
func=spi.Rbf(xData,yData,smooth=.5)
xPlot=np.linspace(xData.min(),xData.max(),num=1000)
fwhmMin=np.round(fit(xPlot,*params).min(),1)
fwhmX=np.round(xPlot[np.argmin(fit(xPlot,*params))],1)
plt.title('minVal= '+str(fwhmMin)+' mm, minPoint= '+str(fwhmX)+' mm')
plt.plot(xPlot,fit(xPlot,*params),c='r',label='data')

#plt.plot(xPlot,func(xPlot),c='green',label='rbf')
plt.scatter(xData,yData,label='fit')
plt.ylabel('FWHM, mm')
plt.grid()
plt.legend()
plt.xlabel('relative longitudinal distance from nozzle, mm')
plt.savefig('boobs')
plt.show()
```
